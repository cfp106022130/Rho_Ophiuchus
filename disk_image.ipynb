```python
from astropy.io import fits
from astropy.wcs import WCS
import numpy as np
import matplotlib.pyplot as plt
from astropy.coordinates import SkyCoord
from astropy import units as u
from astropy.nddata import Cutout2D
import matplotlib.colors as colors
import matplotlib.font_manager as fm
from mpl_toolkits.axes_grid1.anchored_artists import AnchoredSizeBar



filename = 'member.uid___A001_X8c5_X56.DoAr_25_sci.spw0_1_2_3.mfs.I.manual.image.pbcor.fits'
hdu = fits.open(filename)[0]
wcs = WCS(hdu.header)
data = hdu.data[0][0]

wcs = wcs.dropaxis(dropax=2)    # drop the 'FREQ' WCS axis
wcs = wcs.dropaxis(dropax=2)    # drop the 'STOKES' WCS axis

fig = plt.figure()
ax = fig.add_subplot(221, projection=wcs)
ax.set_xlabel(' ', fontsize='small')
ax.set_ylabel(' ', fontsize='small')
ax.tick_params(axis='both', which='major', labelsize=5)

center =  SkyCoord('16h26m23.6819s -24d43m14.3076s', frame='icrs')
size = u.Quantity((2, 2), u.arcsec)
cutout = Cutout2D(data, center, size, wcs=wcs)
ax.imshow(cutout.data, origin='lower', vmax = 0.00165, vmin = -0.0002
          , norm = colors.PowerNorm(gamma=0.8)) 
fontprops = fm.FontProperties(size=5)
pix = hdu.header['CDELT1']*3600                  # in arcsecond
size = 0.5/pix
scalebar = AnchoredSizeBar(ax.transData, size, str(round(0.5*1/7.22210*10**3,0))+'AU', 'lower right', color='white',
                           pad=0.1, frameon = False, fontproperties=fontprops)
ax.add_artist(scalebar)
ax.text(80, 60, 'DoAr 25', style='italic', fontsize=7, color='white')
print(cutout.data.max())
print(cutout.data.min())


filename = 'member.uid___A001_X133d_X4d0.ISO-Oph_196_sci.spw19_21_23_25.cont.I.pbcor.fits'
hdu = fits.open(filename)[0]
wcs = WCS(hdu.header)
data = hdu.data[0][0]

wcs = wcs.dropaxis(dropax=2)    # drop the 'FREQ' WCS axis
wcs = wcs.dropaxis(dropax=2)    # drop the 'STOKES' WCS axis

ax = fig.add_subplot(222, projection=wcs)
ax.set_xlabel(' ', fontsize='small')
ax.set_ylabel(' ', fontsize='small')
ax.tick_params(axis='both', which='major', labelsize=5)

center =  SkyCoord('16h28m16.5048s -24d36m58.4713s', frame='icrs')
size = u.Quantity((2, 2), u.arcsec)
cutout = Cutout2D(data, center, size, wcs=wcs)
ax.imshow(cutout.data, origin='lower', vmax = 0.00165, vmin = -0.0002
          , norm = colors.PowerNorm(gamma=0.8)) 

fontprops = fm.FontProperties(size=5)
pix = hdu.header['CDELT1']*3600                  # in arcsecond
size = 0.5/pix
scalebar = AnchoredSizeBar(ax.transData, size, str(round(0.5*1/7.29480*10**3,0))+'AU', 'lower right', color='white',
                           pad=0.1, frameon = False, fontproperties=fontprops)
ax.add_artist(scalebar)
ax.text(37, 25, 'WSB 60', style='italic', fontsize=7, color='white')
print(cutout.data.max())
print(cutout.data.min())


filename = 'member.uid___A001_X133d_X4d4.s12_0.ISO-Oph_17_sci.spw19_21_23_25.cont.I.manual.image.pbcor.fits'
hdu = fits.open(filename)[0]
wcs = WCS(hdu.header)
data = hdu.data[0][0]

wcs = wcs.dropaxis(dropax=2)    # drop the 'FREQ' WCS axis
wcs = wcs.dropaxis(dropax=2)    # drop the 'STOKES' WCS axis

ax = fig.add_subplot(223, projection=wcs)
ax.set_xlabel('Right Ascension', fontsize='small')
ax.set_ylabel('Declination', fontsize='small')
ax.tick_params(axis='both', which='major', labelsize=5)

center =  SkyCoord('16h26m10.3315s -24d20m55.3s', frame='icrs')
size = u.Quantity((2, 2), u.arcsec)
cutout = Cutout2D(data, center, size, wcs=wcs)
ax.imshow(cutout.data, origin='lower', vmax = 0.00165, vmin = -0.0002
          , norm = colors.PowerNorm(gamma=0.8)) 
fontprops = fm.FontProperties(size=5)
pix = hdu.header['CDELT1']*3600                  # in arcsecond
size = 0.5/pix
scalebar = AnchoredSizeBar(ax.transData, size, str(0.5*140)+'AU', 'lower right', color='white',
                           pad=0.1, frameon = False, fontproperties=fontprops)
ax.add_artist(scalebar)
ax.text(37, 25, 'GSS 26', style='italic', fontsize=7, color='white')
print(cutout.data.max())
print(cutout.data.min())


filename = 'member.uid___A001_X133d_X4dc.DoAr_44_sci.spw19_21_23_25.mfs.I.manual.image.pbcor.fits'
hdu = fits.open(filename)[0]
wcs = WCS(hdu.header)
data = hdu.data[0][0]

wcs = wcs.dropaxis(dropax=2)    # drop the 'FREQ' WCS axis
wcs = wcs.dropaxis(dropax=2)    # drop the 'STOKES' WCS axis

ax = fig.add_subplot(224, projection=wcs)
ax.set_xlabel(' ', fontsize='small')
ax.set_ylabel(' ', fontsize='small')
ax.tick_params(axis='both', which='major', labelsize=5)

center =  SkyCoord('16h31m33.4567s -24d27m37.5458s', frame='icrs')
size = u.Quantity((2, 2), u.arcsec)
cutout = Cutout2D(data, center, size, wcs=wcs)
ax.imshow(cutout.data, origin='lower', vmax = 0.00165, vmin = -0.0002
        , norm = colors.PowerNorm(gamma=0.8)) 
fontprops = fm.FontProperties(size=5)
pix = hdu.header['CDELT1']*3600                  # in arcsecond
# distance: 140 pc, 1 AU = 1/140 arcsec
size = 0.5/pix
scalebar = AnchoredSizeBar(ax.transData, size, str(round(0.5*1/6.8535*10**3,0))+'AU', 'lower right', color='white',
                           pad=0.1, frameon = False, fontproperties=fontprops)
ax.add_artist(scalebar)
ax.text(50, 35, 'DoAr 44', style='italic', fontsize=7, color='white')
print(cutout.data.max())
print(cutout.data.min())

plt.subplots_adjust(left=0.125, bottom=0.1, right=0.9, top=0.9, wspace=-0.2, hspace=None)
plt.savefig('disks.pdf', format='PDF', transparent=True)
```
